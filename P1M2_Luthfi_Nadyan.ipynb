# B. Problem Background and Objectives
___

In competitive e-commerce environment, understanding the key drivers of profitability is essential for sustaining growth and mantaining a competitive edge. Profitability is not determined by a single factor, but rather by a complex interaction of variables that span across customer behavior, operational logistics, and product performance. By analyzing these variables, businesses can make informed decisions to enhance both short-term gains and long-term strategic outcomes. One major factor influencing profit is customer segmentation. Different customer groups—such as corporate clients, home offices, or individual consumers—tend to exhibit distinct purchasing behaviors. Recognizing these patterns allows businesses to tailor their offerings, pricing strategies, and promotional campaigns to better suit each segment’s needs and potential value.

Shipping modes also play a significant role in profitability. While premium shipping options may increase customer satisfaction, they can also inflate operational costs. Balancing speed, reliability, and cost is critical in determining the overall impact on the bottom line. Likewise, product categories and sub-categories may perform differently depending on the region, reflecting varied consumer preferences and market dynamics across geographic areas. Additionally, order size and cost of goods sold are directly tied to profitability. Larger order quantities may benefit from economies of scale, while high product costs can quickly reducing profit margins. Understanding the relationship between these variables allows companies to identify which products are underperforming or overpriced, and which ones offer the best return.

By leveraging historical sales data to predict profit through regression modeling, businesses can unlock a range of strategic advantages. These include optimizing inventory planning to ensure high-demand items are always in stock, customizing marketing initiatives to suit regional or segment-specific behavior, and evaluating logistics performance to streamline operations and reduce delivery-related costs. In sum, a data-driven approach to profit prediction empowers companies to operate more efficiently and profitably in a highly competitive digital marketplace.

## 2. Correlation of Profit to Sales, Quantity, and Cost

```python
# Check the skewness
numeric_cols = df_new.select_dtypes(exclude='object')
numeric_cols = numeric_cols.drop(['Order_ID', 'Order_Date'], axis=1)
skew =[]
for col in numeric_cols:
    skew.append([col, round(numeric_cols[col].skew(),1), np.where(
    (round(numeric_cols[col].skew(),2) <= 0.5) & (round(numeric_cols[col].skew(),2) >= -0.5),
    'Normal','Skewed')])

skewness = pd.DataFrame(columns=['Features', 'Skewness', 'Distribution'], data= skew)
skewness
```

Because the `Profit`, `Sales`, `Quantity`, and `Cost` are skewed I'm going to do the correlation using spearman correlation.

```python
# Create numerical columns
df_corr = numeric_cols

# Initialize the corr heatmap
corr_mat = df_corr.corr(method='spearman')

# Visualize the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Correlation')
plt.show()
corr_mat
```

| Features Corr. | Correlation Value | Intepretation |

| Cost - Profit | 0.44 | Moderate Correlation and Positive |
| Profit - Sales | 0.82 | Strong Correlation and Positive |
| Quantity - Sales | 0.41 | Moderate Correlation and Positive |
| Quantity - Profit | 0.33 | Weak Correlation and Positive |
| Quantity - Cost | 0.35 | Weak Correlation and Positive |

Business Insight:

1. Cost has a stronger correlation with Sales (0.83) than with Profit (0.44).

  - This indicates that as Cost increases, Sales tend to increase significantly, suggesting that higher operational or marketing spending maybe contributing to more sales volume as it shows on the graph.

  - However, Profit is only moderately correlated with Cost, meaning that not all increased Cost leads to increased Profit. Some of the additional Cost may not convert efficiently into bottom-line gains due to factors like discounts, returns, or operational inefficiencies. Which means, it indicates that if company generate more sales and make some profit it will increase the cost either but not that significant.

2. Profit is strongly correlated with Sales (0.82).

  - A higher Sales figure tends to directly relate to higher Profit, which is expected in most businesses.

  - This suggests that strategies to increase sales volume may effectively improve profitability, assuming costs are controlled (back to the correlation of cost with sales it directly positive).

3.  Quantity has a weaker relationship with both Profit (0.33) and Sales (0.41).

  - Selling more units does not always guaranteed increase profit and sales proportionally.

  - This might implied the discount or even promotional campaign that company applied doesn't seem to scale profitablity even more.

## 4. Sales and Revenue Based on Country, Region, and State

```python
# Groupby Region by Performance
region_performance = df_new.groupby('Region').agg(
    tot_sales = ('Sales', 'sum'),
    tot_profit = ('Profit', 'sum')
).reset_index()

# Use pd.melt for grouped barplot (melt is use to merge the bar plot using one benchmark value in this cases is 'Country')
region_melted = pd.melt(
    region_performance,
    id_vars=['Region'], # Unpivot Column
    value_vars=['tot_sales', 'tot_profit'], # Column that combined
    var_name='Comparison Metrics', # New Column named metrics
    value_name='Value' # Show value
)

# Profit Margin
region_performance['Profit_Margin %'] = round((region_performance['tot_profit'] / region_performance['tot_sales']) * 100, 2)

# Visualize
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='Region',
    y='Value',
    hue='Comparison Metrics',
    data=region_melted,
    palette='pastel'
)

# Setting the position of the annot
for bar in ax.patches:
    y_val = bar.get_height()    # value for y setting height position
    x_pos = bar.get_x() + bar.get_width() / 2.0  # half of the width settings
    label = f"{int(y_val)}" # label using the value from the group data
    if y_val >= 0:
        ver_position = 'bottom'
    else:
        ver_position = 'top'
    plt.text(x_pos, y_val, label, ha='center', va=ver_position, fontsize=9, weight='bold')
 
plt.title('Comparison of Sales vs Profit in each Region')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales vs Profit (Value)', fontsize=12)
plt.legend(title='Metrics')
plt.show()
region_performance
```

### b. Categorical Columns Significancy (ANOVA one way test)

To test the relationship significance between Categorical Feature and Numerical Target (Profit), ANOVA is suitable for identifying significancy of such features to target. ANOVA will identified the meaning of the mean of features to the profit and test the relationship of the two.

Hypothesis:

- H0 : Mean within group of feature not significance to the target.
- H1 : Mean within group of feature significance to the target.

```python
# Store new variable just for cat cols
X_train_justtest2 = X_train[categorical_cols]

# Using One Way ANOVA Test for categorical
f_stat = []
p_val = []
cols1 = []
interpretation = []

for col in X_train_justtest2.columns:
    groups = [
        y_train[X_train_justtest2[col] == group]
        for group in X_train_justtest2[col].unique()
    ]
    
    if len(groups) > 1:
        f_statistics, p_value = f_oneway(*groups)

        cols1.append(col)
        p_val.append(p_value)
        f_stat.append(f_statistics)
        
        if p_value < 0.05:
            interpretation.append('Significant Relationship')
        else :
            interpretation.append('No Significant Relationship')

pd.DataFrame({
    'Features': cols1,
    'F-Statistics': f_stat,
    'P-Values': p_val,
    'Results': interpretation
})
```

## 7. Build the Pipeline

```python
# Separating Categorical Variables
cfe_cols = ['State']
tar_cols = ['Order_Year', 'Order_Month', 'Order_DayOfWeek']
categorical_cols = [col for col in categorical_cols if col not in cfe_cols]
categorical_cols = [col for col in categorical_cols if col not in tar_cols]

print('Count Encoder Columns: ', cfe_cols)
print('Target Encoder Columns: ', tar_cols)
print('Categorical Column: ', categorical_cols)
print('Numerical Column: ', numeric_cols)
```

```python
# Numerical Pipeline
num_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', RobustScaler())
])

# Pipeline with One Hot Encoder
cat_transformer_ohe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(sparse_output=False))
])

# Pipeline for categorical columns through Count Frequency Encoder
cat_transformer_cfe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', CountEncoder(normalize=True, handle_missing='value', handle_unknown='value'))
])

# Pipeline for categorical columns through Target Encoder
cat_transformer_tar = Pipeline(steps=[
    ('encoder', TargetEncoder(handle_missing='value', handle_unknown='value'))
])

# Pipeline untuk preprocessing dataset 2
preprocessor = ColumnTransformer(transformers=[
    ('numerical', num_transformer, numeric_cols),
    ('categorical', cat_transformer_ohe, categorical_cols),
    ('categorical1', cat_transformer_cfe, cfe_cols),
    ('categorical2', cat_transformer_tar, tar_cols)
])
```

Pipeline Explanation:

- Numerical Features: From prior analysis that addresed and identified high skewness and presence of outliers in features like `Cost`, `Sales`, and `Quantity` this project use Robust Scaler because such scaler is superior to other scalers because it uses the median and interquartile range, making it immune to the influence of the sensitivity of extreme values. This ensure that outliers do not extort the feature space, allowing the model to learn from the true distribution of the dataset. Using KNN Imputer is crucial because it count the distance between the most close to the other features by the euclidian distance it means the imputer really read the pattern of other features to impute it.

- For Time-Based and Seasonality features using Target Encoder on feature like `Order_Year`, `Order_Month`, `Order_DayOfWeek` for model to learn about representation their historical average profit and advantages of avoiding high dimentionality curse like the journal - [Efficient Representations for High-Cardinality
Categorical Variables in Machine Learning](https://drive.google.com/drive/folders/1tBiZZlOJYqTruWz88ml0Ja0htdEx2y2b). This inject a very strong empirically to the historical data that happened, and explicitly shows the machine to learn about which years, and months are historically high and low performers

- Target Encoder example:

    Suppose we have like year for the example (e.g. 2017, and 2018) then,

    Calculate Averages: For each year (e.g. "2017", "2018), calculate the average of the target variable (Profit) from the training data only.

    - 2017 Average Profit = 100
    
    - 2018 Average Profit = 55

    Replace Categories: Swap the year categories with their calculated averages. The feature like Year in this case ['2017', '2018'] will shows values of [100, 55], respectively.

- For high cardinality features applying `CountEncoder` to `State` (with its 127 unique value) is a good choice since the frequency can of the value shows could replaced the information and avoiding high dimentionality data. Instead using One-Hot Encoder, for this particular column that could create high sparcity of column, I'm going to use count encoder to still preserving and representing the column's information value of each state. This could lead to captures the 'importance' of a state in terms of transactions volume without using the target variable, thus completely avoiding data leakage while managing high dimentionality. One of the advantages of CountEncoder too is to avoid any missing values to the models after preprocessing. In this process, I use simple imputer with strategy most frequent with assumption of state not change overtime, and I want consistent preprocessing.

- Frequency Encoder (CountEncoder) example:

    Suppose we want to only count for `Region` category:

    - Central: Appears 5 times -> Frequency (with normalize=True) = 5 = 5 / 10 = 0.5
    
    - North: Appears 2 times -> Frequency (with normalize=True) = 2 / 10 = 0.2

    - South: Appears 2 times -> Frequency (with normalize=True) = 2 / 10 = 0.2

    Thus the learn encoding would become the percentage of (0.5, 0.2, 0.2)

- For relatively low dimentionality categorical features `One-Hot Encoder` is the safe and standards choice. It makes no assumptions about order and allows the model to learn an independent weight for each category, which suitable and perfect normalization for features with few unique values.

# H. Model Definition

Such description of the model uses in this project I'm going to present it in the form of table:

| Model | Functionality | Advantages | Disadvantages |
| --- | --- | --- | --- |
| K-Nearest Neighbors (KNN) Regressor | Predicts by averaging the target values of the 'K' most similar data points (neighbors) from the training set | Non-parametric (no data distribution assumptions), Good for non-linear data | Sensitive to irrelevant features (curse of dimensionality), Computationally expensive at prediction time, Requires careful feature scaling |
| Support Vector Regressor (SVR) | Finds a hyperplane that best fits the data by keeping as many points as possible within a specified margin, tolerating some errors | Effective in high-dimensional spaces. | Less intuitive to interpret, Sensitive to hyperparameter tuning, Can be slow on large datasets | 
| Decision Tree Regressor | Learns a set of hierarchical if-else rules to partition the data, making predictions based on the average target value of the final leaf node | Highly interpretable, Handles numerical and categorical data well, Captures non-linear relationships and feature interactions automatically | Prone to overfitting, Can be unstable | 
| Random Forest Regressor | An ensemble method that builds many Decision Trees and averages their predictions to make a more robust forecast | Reduces overfitting significantly, High accuracy out-of-the-box, Provides feature importance scores | Less interpretable, Can be slow to train and requires more memory | 
| Gradient Boosting Regressor | An ensemble method that builds trees sequentially, where each new tree is trained to correct the errors of the previous ones |  Often provides the highest predictive accuracy, Flexible and highly tunable, Handles different types of data well | Prone to overfitting if not tuned carefull, Computationally expensive and slower to train |

## 7. Hyperparameter Tuning: Find the Best Model

### a. Setting Parameter

```python
# Parameter Random Forest
params = {
          'regressor__max_leaf_nodes': [None, 1, 2, 3],
          'regressor__max_depth': [None, 1, 2],                 
          'regressor__min_samples_leaf': [1, 2, 3, 4, 5, 6, 7, 8],        
          'regressor__random_state': [42]
         }
```

This is the parameters that I set for the Random Forest Regressor during the hyperparameter tuning later. 

- Regressor max leaf nodes: best nodes are defined as relative reduction in impurity

- Regressor max depth: the maximum depth of tree to split by default it has paramater (None).

- Regressor min samples leaf: The minimum number of samples required to be at a leaf node (smooth the model)

- Random state = 42

I will only keep crucial parameter just like max depth and minimal samples leaf to be in the parameters to prevent it from overcomplicated parameters to the model that lead to non-optimal parameter tuning.

```python
# Grid Search CV for tuning hyperparameter
grid_cv = GridSearchCV(
    estimator=RF_model,
    param_grid=params,
    scoring='neg_mean_absolute_error',
    cv=5,
    n_jobs=-1,
    verbose=1
)
```

- Scoring in this section will focused to minimize the MAE metrics even more, therefore the scoring parameter on focusing to reduce the MAE.

```python
# Fit to train the data
grid_cv.fit(X_train, y_train)
```

```python
# Display the best parameters
grid_cv.best_params_
```

This is the best parameter after tuning to use with Random Forest Regressor Model. I choose to using Grid CV because it gives consistent output on the parameter and does not change overtime when the code in this notebook run.

```python
# Insert the best parameter to the variable
best_model = grid_cv.best_estimator_
```
